```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

# hepatitis

```python
df = pd.read_csv('HepatitisCdata.csv')
```

```python
df.head()
```

```python
df['Category'].unique()
```

```python
df["Category"].replace({
    "0=Blood Donor": 1, 
    "0s=suspect Blood Donor": 2,
    "1=Hepatitis": 3,
    "2=Fibrosis": 4,
    "3=Cirrhosis": 5
}, inplace=True)
df["Sex"].replace({
    "m": True,
    "f": False
}, inplace=True)
df = df.drop(['Unnamed: 0'], 1)
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.reset_index()
```

```python
df.info()
```

```python
training_data, testing_data = train_test_split(df, test_size=0.2, random_state=25, shuffle=True)
```

```python
training_data.to_csv('hepatitis_train.csv', index=False)
```

```python
testing_data.to_csv('hepatitis_test.csv', index=False)
```

# Thyroid

```python
df = pd.read_csv('thyroid-dataset.csv')
```

```python
df.info()
```

```python
df = df.replace('?', np.nan)
# df = df.dropna()
```

```python
df.info()
```

```python
df = df.drop(['TBG'], 1)
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
df.head()
```

```python
df["sex"].replace({"F": True, "M": False}, inplace=True)
df["on thyroxine"].replace({"t": True, "f": False}, inplace=True)
df["query on thyroxine"].replace({"t": True, "f": False}, inplace=True)
df["on antithyroid medication"].replace({"t": True, "f": False}, inplace=True)
df["sick"].replace({"t": True, "f": False}, inplace=True)
df["pregnant"].replace({"t": True, "f": False}, inplace=True)
df["thyroid surgery"].replace({"t": True, "f": False}, inplace=True)
df["I131 treatment"].replace({"t": True, "f": False}, inplace=True)
df["query hypothyroid"].replace({"t": True, "f": False}, inplace=True)
df["query hyperthyroid"].replace({"t": True, "f": False}, inplace=True)
df["lithium"].replace({"t": True, "f": False}, inplace=True)
df["goitre"].replace({"t": True, "f": False}, inplace=True)
df["tumor"].replace({"t": True, "f": False}, inplace=True)
df["hypopituitary"].replace({"t": True, "f": False}, inplace=True)
df["psych"].replace({"t": True, "f": False}, inplace=True)
df["TSH measured"].replace({"t": True, "f": False}, inplace=True)
df["T3 measured"].replace({"t": True, "f": False}, inplace=True)
df["TT4 measured"].replace({"t": True, "f": False}, inplace=True)
df["T4U measured"].replace({"t": True, "f": False}, inplace=True)
df["FTI measured"].replace({"t": True, "f": False}, inplace=True)
df["TBG measured"].replace({"t": True, "f": False}, inplace=True)
df["referral source"].replace({'SVHC': 1, 'SVI': 2, 'other': 3, 'SVHD': 4, 'STMW': 5}, inplace=True)
df["binaryClass"].replace({'P': True, 'N': False}, inplace=True)








```

```python
df.head()
```

```python
training_data, testing_data = train_test_split(df, test_size=0.2, random_state=42, shuffle=True)
training_data.to_csv('thyroid_train.csv', index=False)
training_data.to_csv('thyroid_test.csv', index=False)
```

```python
testing_data['binaryClass'].value_counts()
```
